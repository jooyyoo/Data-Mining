```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file = "C:/Users/USER/Downloads/London.csv"
df = pd.read_csv(file)
df = df[['House Type','Price','Area in sq ft']]
df.set_index(keys='House Type',inplace=True)
df_sort= df.sort_values(by="House Type")


df_sort_= df_sort.loc[["Bungalow"],:]
sns.regplot(x="Price",y="Area in sq ft",data = df_sort_)
plt.title("Bungalow")
plt.show()

df_sort_ = df_sort.loc[["Duplex"],:]
sns.regplot(x="Price",y="Area in sq ft",data = df_sort_)
plt.title("Duplex")
plt.show()

df_sort_= df_sort.loc[["Flat / Apartment"],:]
sns.regplot(x="Price",y="Area in sq ft",data = df_sort_)
plt.title("Falt / Apartment")
plt.show()

df_sort_= df_sort.loc[["House"],:]
sns.regplot(x="Price",y="Area in sq ft",data = df_sort_)
plt.title("House")
plt.show()

df_sort_= df_sort.loc[["Mews"],:]
sns.regplot(x="Price",y="Area in sq ft",data = df_sort_)
plt.title("Mews")
plt.show()

df_sort_= df_sort.loc[["New development"],:]
sns.regplot(x="Price",y="Area in sq ft",data = df_sort_)
plt.title("New development")
plt.show()

df_sort_ = df_sort.loc[["Penthouse"],:]
sns.regplot(x="Price",y="Area in sq ft",data = df_sort_)
plt.title("Penthouse")
plt.show()

df_sort_= df_sort.loc[["Studio"],:]
sns.regplot(x="Price",y="Area in sq ft",data = df_sort_)
plt.title("Studio")
plt.show()
```
