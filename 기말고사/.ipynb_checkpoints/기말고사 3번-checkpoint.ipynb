```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file = "C:/Users/USER/Downloads/London.csv"
df = pd.read_csv(file)
df = df[['House Type','Price','Area in sq ft']]
df.set_index(keys='House Type',inplace=True)
df_sort= df.sort_values(by="House Type")


df_sort_Bungalow = df_sort.loc[["Bungalow"],:]
df_sort_Bungalow
sns.regplot(x="Price",y="Area in sq ft",data = df_sort_Bungalow)
plt.show()

df_sort_Bungalow = df_sort.loc[["Duplex"],:]
df_sort_Bungalow
sns.regplot(x="Price",y="Area in sq ft",data = df_sort_Bungalow)
plt.show()

df_sort_Bungalow = df_sort.loc[["House"],:]
df_sort_Bungalow
sns.regplot(x="Price",y="Area in sq ft",data = df_sort_Bungalow)
plt.show()

df_sort_Bungalow = df_sort.loc[["Mews"],:]
df_sort_Bungalow
sns.regplot(x="Price",y="Area in sq ft",data = df_sort_Bungalow)
plt.show()

df_sort_Bungalow = df_sort.loc[["New Development"],:]
df_sort_Bungalow
sns.regplot(x="Price",y="Area in sq ft",data = df_sort_Bungalow)
plt.show()

df_sort_Bungalow = df_sort.loc[["Penthouse"],:]
df_sort_Bungalow
sns.regplot(x="Price",y="Area in sq ft",data = df_sort_Bungalow)
plt.show()

df_sort_Bungalow = df_sort.loc[["Studio"],:]
df_sort_Bungalow
sns.regplot(x="Price",y="Area in sq ft",data = df_sort_Bungalow)
plt.show()
```

```python
#House Type 별로 집가격과 면적 사이간의 유효 범위가 포함된 선형회귀 plt 도출
df1 = df['House Type'].value_counts()
df1 = pd.DataFrame(df1)
df1
```

```python
df2 = df.groupby('House Type')['Price','Area in sq ft'].mean()
new_df = pd.concat([df1,df2], axis = 1)
new_df
```
