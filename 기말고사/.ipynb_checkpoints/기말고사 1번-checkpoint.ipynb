```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

file = "C:/Users/USER/Downloads/nile.csv"
df.columns = ['year','flood']

df_min = df['flood'].min()
df_max = df['flood'].max()
df_mean = df['flood'].mean() #평균
df_std = df['flood'].std() #표준편차
df_lins = np.linspace(df_min,df_max, 100)
df_norm = stats.norm.pdf(df_lins, df_mean, df_std)

plt.hist(df['flood'],density=1)
y = (1 / np.sqrt(2 * np.pi * df_std**2)) * np.exp(-(df_lins-df_mean)**2 / (2 * df_std**2))
plt.plot(df_lins, y)
plt.show()
```
