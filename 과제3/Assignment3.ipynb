```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc, font_manager

font_path="c:/Windows/Fonts/NGULIM.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font_name)

file = "C:/student_health_2.csv"
df = pd.read_csv(file, encoding='cp949')

#학년만 따로 추출하여 df1이라는 새로운 dataframe으로 만들어준다
df1 = df['학년'].value_counts(sort=False)
df1 = pd.DataFrame(df1)

#gruopby , agg 함수를 통해 학년별 키 몸무게 수축기 이완기를 평균 내어 준다
#grouped 의 type은 pandas.core.frame.DataFrame, 즉 새로운 DataFrame 생성
grouped = df.groupby('학년')['키','몸무게','수축기','이완기'].agg(['mean'])

#concat 함수를 사용하여 df1, grouped 를 합쳐준 새로운 DataFrame을 만들어준다. new_df
#칼럼명을 알아보기 쉽게 변셩해준다
new_df = pd.concat([df1,grouped], axis = 1)
new_df.columns = ['학년 인원수','키 평균', '몸무게 평균', '수축기 평균', '이완기 평균']


'''
pyplot 은 시퀀스 스칼라를 지원한다. 
즉, "순서가 있는 숫자"로써 리스트(list) 또는 넘파이의 series 같은 데이터 유형을 의미한다
따라서 x,y 축에 플롯하고 싶은 데이터를 list 로 지정해야 한다.
'''
'''학년 수'''
xs = new_df.index.to_list()
ys = new_df['학년 인원수'].to_list()
plt.xlabel('학년')
plt.ylabel('학년별 인원수')
plt.bar(xs,ys)
plt.show()

'''평균 키'''
xs = new_df.index.to_list()
ys = new_df['키 평균'].to_list()
plt.xlabel('학년')
plt.ylabel('키')
plt.bar(xs,ys)
plt.show()

'''평균 몸무게'''
xs = new_df.index.to_list()
ys = new_df['몸무게 평균'].to_list()
plt.xlabel('학년')
plt.ylabel('몸무게')
plt.bar(xs,ys)
plt.show()

'''평균 수축기'''
xs = new_df.index.to_list()
ys = new_df['수축기 평균'].to_list()
plt.xlabel('학년')
plt.ylabel('수축기')
plt.xlim([0,6])
plt.bar(xs,ys)
plt.show()

'''평균 이완기'''
xs = new_df.index.to_list()
ys = new_df['이완기 평균'].to_list()
plt.xlabel('학년')
plt.ylabel('이완기')
plt.xlim([0,6])
plt.bar(xs,ys)
plt.show()
```
