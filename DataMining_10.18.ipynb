```python
df1 = pd.DataFrame({'key':['A','B','C','D'],'value':np.random.rand(4)})
df2 = pd.DataFrame({'key':['B','D','D','E'],'value':np.random.rand(4)})
mg_r = pd.merge(df1,df2,on='key')
mg_r
```

```python
mg_r = pd.merge(df1,df2,on='key',how="left")
mg_r
```

```python
mg_r = pd.merge(df1,df2,on='key',how="right")
mg_r
```

```python
mg_r = pd.merge(df1,df2,on='key',how='outer')
mg_r
```

```python
mg_r = pd.merge(df1,df2,on='key',how='inner')
mg_r
```

```python
left = pd.DataFrame({'key1': ['Z0','Z0','Z1',"Z2"],
                    'key2': ['Z0','Z1','Z0','Z1'],
                    'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key1': ['Z0','Z0','Z1',"Z2"],
                    'key2': ['Z0','Z0','Z0','Z0'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})

result = pd.merge(left,right,on=['key1','key2'],how = 'inner')
result
```
