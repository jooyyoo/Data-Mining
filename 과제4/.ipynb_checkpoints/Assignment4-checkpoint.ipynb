```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

file = "C:/student_health_3.csv"
df = pd.read_csv(file, encoding = 'CP949')

df1 = df[['수축기','이완기','키','몸무게']]
df2 = df['학년']

X_train, X_test, y_train, y_test = train_test_split(df1,df2,random_state=0)

k_list = range(1,101)
acc = []

for k in k_list:
    knn = KNeighborsClassifier(n_neighbors = k) #모델 생성
    knn.fit(X_train, y_train)#모델 학습
    acc.append(knn.score(X_test,y_test))
    
plt.plot(k_list,acc)
plt.xlabel("k")
plt.ylabel("accuracy")
plt.title("knn")
plt.show()

print("KNN")
print("학습세트 정확도: ",knn.score(X_train,y_train))
print("테스트세트 정확도: ",knn.score(X_test,y_test))

print("======================================")

l = ['수축기','이완기','키','몸무게']
def re(li):
    for i in range(len(df[li])):
        if df1[li].mean() <= df1[li][i]:
            df1[li][i] = 1
        else:
            df1[li][i] = 0
        
for i in l:
    re(i)

#데이터 세트 준비하기
features = df1
grade = df2

#학습세트/평가세트 분리하기
train_features, test_features, train_labels, test_labels = train_test_split(features, grade)

#데이터 정규화하기
scaler = StandardScaler()
train_features = scaler.fit_transform(train_features)
test_features = scaler.transform(test_features)
#학습 세트로 fit 을 해주었기 떄문에 평가세트에서는 별도로 fit 필요 없다

#모델 생성 및 평가하기
model = LogisticRegression()
model.fit(train_features,train_labels)

print("LogisticRegression")
print("학습세트 정확도: ",model.score(train_features, train_labels))
print("테스트세트 정확도: ",model.score(test_features, test_labels))
```
