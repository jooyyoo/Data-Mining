```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

file = "C:/student_health_3.csv"
df = pd.read_csv(file, encoding = 'CP949')

df1 = df[['','','','']]
df2 = df['']

X_train, X_test, y_train, y_test = train_test_split(df1,df2,random_state=0)

k_list = range(1,101)
acc = []

for k in k_list:
    knn = KNeighborsClassifier(n_neighbors = k) # 
    knn.fit(X_train, y_train)# 
    acc.append(knn.score(X_test,y_test))
    
plt.plot(k_list,acc)
plt.xlabel("k")
plt.ylabel("accuracy")
plt.title("knn")
plt.show()

print("KNN")
print(" : ",knn.score(X_train,y_train))
print(" : ",knn.score(X_test,y_test))

print("======================================")

l = ['','','','']
def re(li):
    for i in range(len(df[li])):
        if df1[li].mean() <= df1[li][i]:
            df1[li][i] = 1
        else:
            df1[li][i] = 0
        
for i in l:
    re(i)

#  
features = df1
grade = df2

#/ 
train_features, test_features, train_labels, test_labels = train_test_split(features, grade)

# 
scaler = StandardScaler()
train_features = scaler.fit_transform(train_features)
test_features = scaler.transform(test_features)
#  fit      fit  

#   
model = LogisticRegression()
model.fit(train_features,train_labels)

print("LogisticRegression")
print(" : ",model.score(train_features, train_labels))
print(" : ",model.score(test_features, test_labels))
```
