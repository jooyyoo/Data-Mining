```python
import pandas as pd
file = "C:/testdata.csv"
df = pd.read_csv(file)
del df['Unnamed: 0']
df
```

```python
for column_name in df:
    print(column_name)
```

```python
for column_name in df:
        print(" ")
        print("{}".format(column_name))
        avg = df[column_name].mean()
        std = df[column_name].std()
        maxx = df[column_name].max()
        minn = df[column_name].min()
        quan = df[column_name].quantile([0.15,0.35,0.5,0.65,0.85])
        print("""
        avg : {}
        std : {}
        max : {}
        min : {}
        15% : {}
        35% : {}
        50% : {}
        65% : {}
        85% : {}""".format(avg,std,maxx,minn,quan[0.15],quan[0.35],quan[0.5],quan[0.65],quan[0.85]))
    
```

```python
import numpy as np
dfnumpy = df.to_numpy()
dfnumpy
```

```python
arr = np.array([])
li = ['column1','column2','column3','column4','column5']
```

```python
for i in range(len(li)):
    avg = 0
    print("{}".format(li[i]))
    for j in range(len(dfnumpy)):
        arr = np.append(arr,dfnumpy[j][i])
        avg = arr.mean()
        std = arr.std()
        maxx = arr.max()
        minn = arr.min()
        quan15 = np.quantile(arr,.15)
        quan35 = np.quantile(arr,.35)
        quan50 = np.quantile(arr,.5)
        quan65 = np.quantile(arr,.65)
        quan85 = np.quantile(arr,.85)
    print("""
        avg : {}
        std : {}
        max : {}
        min : {}
        15% : {}
        35% : {}
        50% : {}
        65% : {}
        85% : {}""".format(avg,std,maxx,minn,quan15,quan35,quan50,quan65,quan85))
    print("")
```
